```python
#Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
#Using the school_data_complete_df DataFrame, we can get the average reading score using the following code:
# Calculate the average reading score.
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Calculate the average math score.
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#For math and reading assessment tests in this school district, the passing score was 70. 
#herefore, we need to get all the math and reading scores that are greater than or equal to 70. 
#To do this, in a new cell, assign a passing_math variable to the math_score column in school_data_complete_df, 
#where all the math scores are equal to or greater than 70.

passing_math = school_data_complete_df["math_score"] >= 70
passing_reading = school_data_complete_df["reading_score"] >= 70
```

```python
#To find the passing_math variable, run passing_math in a new cell. 
#The result is Boolean values for the rows, where "True" means the score is equal to or greater than 70, and 
#"False" means the score is not equal to or greater than 70 

passing_math
```

```python
#To get all the students who passed math and all the students who passed reading, 
#we need to filter our school_data_complete_df DataFrame

# Get all the students who are passing math in a new DataFrame.
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()
```

```python
# Get all the students that are passing reading in a new DataFrame.
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading.head()
```

```python
#get the number of students who passed each subject. 
passing_math["student_name"].count()
```

```python
#get the number of students who passed each subject.
passing_reading["student_name"].count()
```

```python
#To get the percentage of students who passed math and reading, 
#divide the passing_math_count and the passing_reading_count by the total number of students, 
#and then multiply by 100.Recall that the student count was calculated using this code:
#school_data_complete_df["Student ID"].count()

#Since we are calculating a percentage, we need to convert the student_count to a number with a decimal, 
#or floating-point decimal, by using float().
```

```python
# Calculate the percent that passed math.
passing_math_percentage = passing_math_count / (student_count) * 100
print(passing_math_percentage)

# Calculate the percent that passed reading.
passing_reading_percentage = passing_reading_count / (student_count) * 100
print(passing_reading_percentage)
```

```python
#To get the overall passing percentage, 
#we need to get all the students who passed both math and reading and divide by the total number of students.

# Calculate the students who passed both math and reading.
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]

passing_math_reading.head()
```

```python
#How would you get the total number of students that passed math and reading?
#passing_math_reading["student_name"].count()
passing_math_reading.student_name.count()
```

```python
# Calculate the number of students who passed both math and reading.
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
overall_passing_math_reading_count
```

```python
# Calculate the overall passing percentage.
overall_passing_percentage = overall_passing_math_reading_count / student_count * 100
overall_passing_percentage
```
